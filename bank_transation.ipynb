```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['date'] = df['TransactionDate'].dt.hour
```

```python
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])
df['Day'] = df['PreviousTransactionDate'].dt.day_name()
df['Day']
```

```python
 print("Total TransactionsDuration:", len(df))
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])
print("Average PreviousTransactionDate:", df['PreviousTransactionDate'].mean())
print("Highest PreviousTranscationDate:", df['PreviousTransactionDate'].max())
print("Lowest PreviousTransaction:",df['PreviousTransactionDate'].min())
print("PreviousTransactionDate count:",df['PreviousTransactionDate'].count())
```

```python
TransactionAmount = df.groupby('TransactionType')['TransactionAmount'].sum()
plt.pie(
    TransactionAmount,
    labels=TransactionAmount.index,
    autopct='%.1f%%',
    startangle=90
)
plt.title('TransactionAmount by TransactionType')
plt.axis('equal')
plt.show()
```

```python
df.groupby('Channel')['TransactionAmount'].sum().sort_values().plot(kind='bar')

plt.title('Total Transaction Amount by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Transaction Amount')
plt.show()
```

```python
sns.histplot(df['TransactionAmount'], bins=30, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
sns.boxplot(x=df['AccountBalance'])
plt.title('Account Balance Distribution (Box Plot)')
plt.show()
```

```python
df.plot(kind='scatter',
               x='PreviousTransactionDate',
               y='TransactionAmount',
               figsize=(8,5),
               color='pink',
               s=80)
plt.title('Previous Transaction Date vs. Transaction Amount')
plt.xlabel('Previous Transaction Date')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.groupby('CustomerOccupation')['TransactionAmount'].sum().sort_values().plot(kind='bar')

plt.title('Total Transaction Amount by CustomerOccupation')
plt.xlabel('CustomerOccupation')
plt.ylabel('Total Transaction Amount')
plt.show()
```

```python
monthly = df.groupby(df['TransactionDate'].dt.to_period('M'))['TransactionAmount'].sum()

monthly.plot(figsize=(10,5), marker='o')
plt.title("Monthly Transaction Trend")
plt.show()
```
